### Question-1
Write a function using only list filter lambda that can tell whether a number is a Fibonacci number or not. You can use a pre-calculated list/dict to store fab numbers till 10000 PTS:100

```
from functools import reduce

n = int(input("Enter a number: "))
fib = lambda n: reduce(lambda x, _: x+[x[-1]+x[-2]], range(n-2),[0,1])

list_fib_10000 = fib(5000)
print(list_fib_10000)
if n in list_fib_10000:
  print("Number-{} is a fibonacci number".format(n))
else:
  print("Number-{} is NOT a fibonacci number".format(n))
```

###Qestion-2:
Using list comprehension (and zip/lambda/etc if required) write five different expressions that: PTS:100
add 2 iterables a and b such that a is even and b is odd
strips every vowel from a string provided (tsai>>t s)
acts like a ReLU function for a 1D array
acts like a sigmoid function for a 1D array
takes a small character string and shifts all characters by 5 (handle boundary conditions) tsai>>yxfn

```
import math

a = [-1,1,2,3,4,-5,0]
s = "indiaby"

# takes a small character string and shifts all characters by 5 (handle boundary conditions) tsai>>yxfn
shift_5 = "".join([chr(ord(i)+5) if ord(i)+5 <= ord('z') else chr(ord('z')) for i in s])
print(shift_5)
```

```
# strips every vowel from a string provided (tsai>>t s)
ovel = "".join([i for i in s if i in ['a','e','i','o','u']])
print(ovel)
```

```
# acts like a ReLU function for a 1D array
relu = [i if i>0 else 0 for i in a]
print(relu)
```

```
# acts like a sigmoid function for a 1D array
sigma = [1/(1+math.exp(-i)) for i in a]
print(sigma)
```

### Question-3:
A list comprehension expression that takes a ~200 word paragraph (write your own paragraph to check), and checks whether it has any of the swear words mentioned in https://github.com/RobertJGabriel/Google-profanity-words/blob/master/list.txt PTS:200

```
para_200words = """hi"""
file1 = open('list.txt', 'r')
Lines = file1.readlines()

swear_words_para = [word for para in para_200words.split(".") for word in para.split("") if word+"\n" in Lines]
print(swear_words_para)
```

### Question-4:

```
import functools

# add only even numbers in a list
a_list = [1,2,3,4,5,6,7,5,4,99,88,76,23]

add_even = functools.reduce(lambda a,b: a+b, [i for i in a_list if i%2==0])
print(add_even)
```

```
# find the biggest character in a string (printable ascii characters)
s = "indiaby"
biggest = functools.reduce(lambda a,b : a if a>b else b, [ord(i) for i in s])
print(chr(biggest))
```

```
# adds every 3rd number in a list
third_num = functools.reduce(lambda a,b : a+b, [a_list[i] for i in range(0,len(a_list),3)])
print(third_num)
```

### Question-5
Using randint, random.choice and list comprehensions, write an expression that generates 15 random KADDAADDDD number plates, where KA is fixed, D stands for a digit, and A stands for Capital alphabets. 10<<DD<<99 & 1000<<DDDD<<9999 PTS:100

```
import random 
import string

# list of alphabets from A to Z
x_alpha = string.ascii_uppercase[:26]
num_plate = ["KA{}{}{}{}".format(random.randint(10,100), random.choice(x_alpha), random.choice(x_alpha), random.randint(1000,10000)) for i in range(15)]
print(num_plate)
```
